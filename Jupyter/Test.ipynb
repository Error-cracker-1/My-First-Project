```python
import tkinter as tk
import webbrowser
from tkinter import N, messagebox

def open_website():
    webbrowser.open("https://error-cracker-1.github.io/My-First-Project")

def say_hello():
    messagebox.showinfo("Welcome", "Hello, Welcome to My project!" + "\n" + "You Are now redirected To My Project Website")
    label.config(text="Hello, Welcome to My project!")

def start_app():
    say_hello()
    open_website()

window = tk.Tk()
window.title("Test App")
window.geometry("300x100")
label = tk.Label(window, text="You are in the Test App!")
label.pack(pady=20)
button = tk.Button(window, text="Start", command=start_app)
button.pack()
window.mainloop()
```

```python
import math
import tkinter as tk
from tkinter import messagebox
import webbrowser

def calculate_sqrt():
    try:
        value = float(entry.get())
        if value < 0:
            raise ValueError("Negative number")
        result = math.sqrt(value)
        messagebox.showinfo("Square Root", f"The square root of {value} is {result}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter a valid non-negative number.")
def open_github():
        webbrowser.open("https://error-cracker-1.github.io/My-First-Project")

window = tk.Tk()
window.title("Square Root Calculator")
window.geometry("320x190")

label = tk.Label(window, text="Enter a number:")
label.pack(pady=(15, 5))

entry = tk.Entry(window, width=20)
entry.pack()

calculate_button = tk.Button(window, text="Calculate", command=calculate_sqrt)
calculate_button.pack(pady=10)

github_button = tk.Button(window, text="Visit My GitHub", command=open_github)
github_button.pack(pady=10)

window.mainloop()
```
