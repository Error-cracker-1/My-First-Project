# MultiFunctional Scientific Calculator

This notebook contains a Tkinter-based calculator app with a simple graphical interface for common arithmetic and scientific operations.

## Features

- Add two numbers
- Subtract the second number from the first number
- Multiply two numbers
- Divide two numbers with zero-division protection
- Calculate powers using the first number as the base and the second number as the exponent
- Calculate the square root of the first number
- Open the project GitHub page with the Support button

## How to Use

1. Run the calculator code cell below.
2. Enter values in the first and second number fields.
3. Select the operation button you want to perform.
4. Read the result in the popup message box.

## Input Notes

- For square root, only the first number field is used.
- For power, the first number is the base and the second number is the exponent.
- Division by zero is blocked and shows an error message.
- Invalid or empty number fields show an error message instead of crashing the app.


```python
import tkinter as tk
import math
import webbrowser
from tkinter import messagebox

themes = {
    "light": {
        "bg": "lightblue",
        "fg": "darkblue",
        "button_bg": "blue",
        "button_fg": "white",
        "active_bg": "blue",
        "active_fg": "white",
    },
    "dark": {
        "bg": "black",
        "fg": "white",
        "button_bg": "gray",
        "button_fg": "white",
        "active_bg": "gray",
        "active_fg": "white",
    },
    "colorful": {
        "bg": "lightyellow",
        "fg": "darkgreen",
        "button_bg": "orange",
        "button_fg": "white",
        "active_bg": "orange",
        "active_fg": "white",
    },
}

def add():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        result = num1 + num2
        messagebox.showinfo("Result", f"The sum is: {result}")
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")
def subtract():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        result = num1 - num2
        messagebox.showinfo("Result", f"The difference is: {result}")
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

def multiply():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        result = num1 * num2
        messagebox.showinfo("Result", f"The product is: {result}")
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")
def divide():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        if num2 == 0:
            messagebox.showerror("Error", "Cannot divide by zero.")
            return
        result = num1 / num2
        messagebox.showinfo("Result", f"The quotient is: {result}")
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")
def exponentiate():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        result = math.pow(num1, num2)
        messagebox.showinfo("Result", f"The result of {num1} raised to the power of {num2} is: {result}")
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")
def square_root():
    try:
        num = float(entry1.get())
        if num < 0:
            messagebox.showerror("Error", "Cannot calculate square root of a negative number.")
            return
        result = math.sqrt(num)
        messagebox.showinfo("Result", f"The square root of {num} is: {result}")
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid number.")
def clear_entries():
    entry1.delete(0, tk.END)
    entry2.delete(0, tk.END)
def open_github():
    webbrowser.open("https://github.com/Error-cracker-1/My-First-Project")

window = tk.Tk()
window.iconbitmap("")
window.configure(bg="lightblue")
window.title("MultiFunctional Calculator")   
window.geometry("420x550")
label = tk.Label(window, text="The Scientific Calculator", font=("Segoe UI", 20, "bold"), bg="lightblue", fg="darkblue")
label.pack(pady=10)
inputs_label = tk.Label(
    window,
    text="Inputs",
    font=("Arial", 12, "bold")
)

inputs_label.pack(pady=5)
divider1 = tk.Label(
    window,
    text="------------------"
)

divider1.pack()
Instruction_label = tk.Label(window, text="Enter 1st Number", font=("Arial", 10))
Instruction_label.pack(pady=5)
entry1 = tk.Entry(window ,bg="white", fg="black", font=("Arial", 12))
entry1.pack(pady=5)
Instruction_label2 = tk.Label(window, text="Enter 2nd Number select an operation:", font=("Arial", 10))
Instruction_label2.pack(pady=5)
entry2 = tk.Entry(window, bg="white", fg="black", font=("Arial", 12))
entry2.pack(pady=5)
operations_label = tk.Label(
    window,
    text="Operations",
    font=("Arial", 12, "bold")
)

operations_label.pack(pady=(15, 10))
button_frame = tk.Frame(window,bg="lightblue")
button_frame.pack(pady=5)
button_add = tk.Button(
    button_frame, 
    text="Add (+)", 
    command=add,
    bg="Blue", 
    fg="white", 
    activebackground="Blue", 
    activeforeground="white",
    width=10,
    font=("Arial", 12, "bold"),
)
button_add.grid(row=0, column=0, padx=5, pady=7)
button_subtract = tk.Button(
    button_frame, 
    text="Subtract (-)",
    command=subtract,
    bg="green", 
    fg="white", 
    activebackground="Green", 
    activeforeground="white", 
    font=("Arial", 12, "bold"), 
    width=10,
)
button_subtract.grid(row=0, column=1, padx=5, pady=7)
button_multiply = tk.Button(
    button_frame, 
    text="Multiply (*)", 
    command=multiply,
    bg="orange", 
    fg="white", 
    activebackground="Orange", 
    activeforeground="white", 
    font=("Arial", 12, "bold"), 
    width=10,
)
button_multiply.grid(row=0, column=2, padx=5, pady=7)
button_divide = tk.Button(
    button_frame, 
    text="Divide (∻)", 
    command=divide,
    bg="red", 
    fg="white", 
    activebackground="Red", 
    activeforeground="white", 
    font=("Arial", 12, "bold"), 
    width=10,
)
button_divide.grid(row=1, column=0, padx=5, pady=7)
button_exponentiate = tk.Button(
    button_frame, 
    text="Power (aⁿ)", 
    command=exponentiate,
    bg="purple", 
    fg="white", 
    activebackground="Purple", 
    activeforeground="white", 
    font=("Arial", 12, "bold"),
    width=10,
)
button_exponentiate.grid(row=1, column=1, padx=5, pady=7)
button_square_root = tk.Button(
    button_frame, 
    text="Root (√)", 
    command=square_root,
    bg="brown", 
    fg="white", 
    activebackground="brown", 
    activeforeground="white", 
    font=("Arial", 12,"bold"), 
    width=10,
)
button_square_root.grid(row=1, column=2, padx=5, pady=7)
button_clear = tk.Button(
    button_frame, 
    text="Clear", 
    command=clear_entries,
    bg="gray", 
    fg="white", 
    activebackground="gray", 
    activeforeground="white",
    font=("Arial", 12, "bold"),
    width=10,
)
button_clear.grid(row=2, column=2, pady=5)
button_github = tk.Button(
    button_frame, 
    text="Support", 
    command=open_github,
    bg="black", 
    fg="white", 
    activebackground="black", 
    activeforeground="white", 
    font=("Arial", 12, "bold"), 
    width=10,
)
button_github.grid(row=2, column=0, columnspan=3, pady=5)
Instruction_label3 = tk.Label(window, text="The Number is Selected From the 1st Number Field For Square Root", font=("Arial", 9), fg="red")
Instruction_label3.pack(pady=5)
Instruction_label4 = tk.Label(window, text="The Base Selected From the 1st Cell and The Exponent From the 2nd Cell", font=("Arial", 9), fg="blue")
Instruction_label4.pack(pady=5)
window.mainloop()
```

# Modern Scrollable Calculator

This is an updated, modern version of the calculator featuring scrollbar support, hover effects, and a theme toggling system (Dark and Light themes).

## Features
- **Modern Theme Colors**: Custom slate dark and nord light color palettes.
- **Interactive Buttons**: Hover effects on all calculation and control buttons.
- **Scrollbar Support**: Uses a Canvas containing a Frame to make the GUI scrollable.
- **Operations**:
  - Add / Subtract Two Numbers
  - Multiply / Divide Two Numbers (with division-by-zero check)
  - Square Root
  - Exponentiate
  - Visit My Website
  - Say Hello
- **Theme Switcher**: Dynamically toggle between Dark and Light themes.


```python
import tkinter as tk
import webbrowser
from tkinter import messagebox

#-------------------Constants-------------------
FONT = ("Segoe UI", 14, "bold")
PROJECT_URL = "https://github.com/Error-cracker-1/My-First-Project"
BUTTON_WIDTH = 20
ENTRY_WIDTH = 20

#-------------------Themes----------------------
THEMES = {
    "dark": {
        "bg": "#1E1E2E",          # Deep Charcoal/Navy (Catppuccin Mocha)
        "fg": "#CDD6F4",          # Off-white Text
        "subtext": "#A6ADC8",     # Muted Text
        "accent": "#89B4FA",      # Sky Blue Accent
        "btn_bg": "#313244",      # Button Background
        "btn_hover": "#45475A",   # Button Hover Background
        "entry_bg": "#181825",    # Entry Background
        "border": "#45475A"
    },
    "light": {
        "bg": "#F2F4F8",          # Soft light gray (Nord-like)
        "fg": "#2E3440",          # Dark text
        "subtext": "#4C566A",     # Muted light text
        "accent": "#5E81AC",      # Dark blue accent
        "btn_bg": "#E5E9F0",      # Soft light gray button
        "btn_hover": "#D8DEE9",   # Hover light gray
        "entry_bg": "#FFFFFF",    # White entry background
        "border": "#D8DEE9"
    }
}

current_theme = "dark"

#-------------------Functions-------------------
def add_numbers():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        result = num1 + num2
        messagebox.showinfo("Result", f"The sum is: {result}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numbers.")

def substract_numbers():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        result = num1 - num2
        messagebox.showinfo("Result", f"The difference is: {result}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numbers.")

def multiply_numbers():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        result = num1 * num2
        messagebox.showinfo("Result", f"The product is: {result}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numbers.")

def divide_numbers():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        if num2 == 0:
            messagebox.showerror("Input Error", "Please enter a non-zero second number.")
        else:
            result = num1 / num2
            messagebox.showinfo("Result", f"The quotient is: {result}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numbers.")

def square_root():
    try:
        num = float(entry1.get())
        if num < 0:
            messagebox.showerror("Input Error", "Please enter a non-negative number.")
        else:
            result = num ** 0.5
            messagebox.showinfo("Result", f"The square root is: {result}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter a valid number.")

def exponentiate():
    try:
        base = float(entry1.get())
        exponent = float(entry2.get())
        result = base ** exponent
        messagebox.showinfo("Result", f"The result of {base} raised to the power of {exponent} is: {result}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numbers.")

def open_website():
    webbrowser.open(PROJECT_URL)

def say_hello():
    messagebox.showinfo("Hello", "Hello, Welcome to My project!")

def update_scroll_region(event=None):
    canvas.configure(scrollregion=canvas.bbox("all"))

#-------------------Theme Engine------------------
def apply_theme():
    t = THEMES[current_theme]
    window.config(bg=t["bg"])
    canvas.config(bg=t["bg"], highlightthickness=0)
    scrollbar_frame.config(bg=t["bg"])

    # Configure Labels
    label.config(bg=t["bg"], fg=t["accent"])
    Instruction_label.config(bg=t["bg"], fg=t["subtext"])
    label1.config(bg=t["bg"], fg=t["fg"])
    label2.config(bg=t["bg"], fg=t["fg"])

    # Configure Entries
    for e in entries:
        e.config(
            bg=t["entry_bg"],
            fg=t["fg"],
            insertbackground=t["fg"],
            highlightbackground=t["border"],
            highlightcolor=t["accent"],
            highlightthickness=1,
            relief="flat"
        )

    # Configure Buttons
    for b in buttons:
        b.config(
            bg=t["btn_bg"],
            fg=t["fg"],
            activebackground=t["btn_hover"],
            activeforeground=t["fg"],
            relief="flat",
            bd=0
        )

#-------------------Theme Toggle system-----------
def toggle_theme():
    global current_theme
    current_theme = "light" if current_theme == "dark" else "dark"
    apply_theme()

#-------------------Main Window-------------------
window = tk.Tk()
window.title("Calculator")
window.minsize(500, 380)

#-------------------Loop system-------------------
labels = []
buttons = []
entries = []

# ---------------- CANVAS ----------------
canvas = tk.Canvas(window, highlightthickness=0)
canvas.pack(side="left", fill="both", expand=True)

# ---------------- SCROLLBAR ----------------
scrollbar = tk.Scrollbar(window, command=canvas.yview)
scrollbar.pack(side="right", fill="y")

canvas.configure(yscrollcommand=scrollbar.set)

# ---------------- FRAME INSIDE CANVAS ----------------
scrollbar_frame = tk.Frame(canvas)
canvas.create_window((0, 0), window=scrollbar_frame, anchor="nw")

scrollbar_frame.bind("<Configure>", update_scroll_region)

# Helper to bind hover effect
def bind_hover(btn):
    btn.bind("<Enter>", lambda e, b=btn: b.config(bg=THEMES[current_theme]["btn_hover"]))
    btn.bind("<Leave>", lambda e, b=btn: b.config(bg=THEMES[current_theme]["btn_bg"]))

# ---------------- CONTENT (GRID LAYOUT ONLY) ----------------
# Row 0: Welcome Label (spans 2 columns)
label = tk.Label(scrollbar_frame, text="The Calculator", font=FONT)
label.grid(row=0, column=0, columnspan=2, padx=10, pady=15, sticky="ew")


# Row 1: Instruction Label (spans 2 columns)
Instruction_label = tk.Label(
    scrollbar_frame, 
    text="Enter numbers in the fields below and click the buttons to perform calculations.", 
    font=("Segoe UI", 10), 
    wraplength=450
)
Instruction_label.grid(row=1, column=0, columnspan=2, padx=10, pady=5, sticky="ew")

# Row 2: Entry 1 Label & Entry 1
label1 = tk.Label(scrollbar_frame, text="First Number:", font=("Segoe UI", 10, "bold"))
label1.grid(row=2, column=0, padx=10, pady=5, sticky="e")

entry1 = tk.Entry(scrollbar_frame, width=ENTRY_WIDTH, font=("Segoe UI", 10))
entry1.grid(row=2, column=1, padx=10, pady=5, sticky="w")
entries.append(entry1)

# Row 3: Entry 2 Label & Entry 2
label2 = tk.Label(scrollbar_frame, text="Second Number:", font=("Segoe UI", 10, "bold"))
label2.grid(row=3, column=0, padx=10, pady=5, sticky="e")

entry2 = tk.Entry(scrollbar_frame, width=ENTRY_WIDTH, font=("Segoe UI", 10))
entry2.grid(row=3, column=1, padx=10, pady=5, sticky="w")
entries.append(entry2)

# Row 4: Buttons (Add / Subtract)
button_add = tk.Button(scrollbar_frame, text="Add Two Numbers", command=add_numbers, width=BUTTON_WIDTH, font=("Segoe UI", 10, "bold"))
button_add.grid(row=4, column=0, padx=10, pady=5)
buttons.append(button_add)
bind_hover(button_add)

button_subtract = tk.Button(scrollbar_frame, text="Subtract Two Numbers", command=substract_numbers, width=BUTTON_WIDTH, font=("Segoe UI", 10, "bold"))
button_subtract.grid(row=4, column=1, padx=10, pady=5)
buttons.append(button_subtract)
bind_hover(button_subtract)

# Row 5: Buttons (Multiply / Divide)
button_multiply = tk.Button(scrollbar_frame, text="Multiply Two Numbers", command=multiply_numbers, width=BUTTON_WIDTH, font=("Segoe UI", 10, "bold"))
button_multiply.grid(row=5, column=0, padx=10, pady=5)
buttons.append(button_multiply)
bind_hover(button_multiply)

button_divide = tk.Button(scrollbar_frame, text="Divide Two Numbers", command=divide_numbers, width=BUTTON_WIDTH, font=("Segoe UI", 10, "bold"))
button_divide.grid(row=5, column=1, padx=10, pady=5)
buttons.append(button_divide)
bind_hover(button_divide)

# Row 6: Buttons (Square Root / Exponentiate)
button_sqrt = tk.Button(scrollbar_frame, text="Square Root", command=square_root, width=BUTTON_WIDTH, font=("Segoe UI", 10, "bold"))
button_sqrt.grid(row=6, column=0, padx=10, pady=5)
buttons.append(button_sqrt)
bind_hover(button_sqrt)

button_exponentiate = tk.Button(scrollbar_frame, text="Exponentiate", command=exponentiate, width=BUTTON_WIDTH, font=("Segoe UI", 10, "bold"))
button_exponentiate.grid(row=6, column=1, padx=10, pady=5)
buttons.append(button_exponentiate)
bind_hover(button_exponentiate)

# Row 7: Buttons (Visit Website / Say Hello)
button_website = tk.Button(scrollbar_frame, text="Visit My Website", command=open_website, width=BUTTON_WIDTH, font=("Segoe UI", 10, "bold"))
button_website.grid(row=7, column=0, padx=10, pady=5)
buttons.append(button_website)
bind_hover(button_website)

button_hello = tk.Button(scrollbar_frame, text="Say Hello", command=say_hello, width=BUTTON_WIDTH, font=("Segoe UI", 10, "bold"))
button_hello.grid(row=7, column=1, padx=10, pady=5)
buttons.append(button_hello)
bind_hover(button_hello)

# Row 8: Toggle Theme
button_theme = tk.Button(scrollbar_frame, text="THEMES", command=toggle_theme, font=("Segoe UI", 10, "bold"))
button_theme.grid(row=8, column=0, padx=10, pady=10, columnspan=2)
buttons.append(button_theme)
bind_hover(button_theme)

apply_theme()
window.mainloop()
```
